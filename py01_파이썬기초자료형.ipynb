```python
# 문자열
print("%s" % '12.34 문자표현')
# 정수
print("%d" % 12.34)
# 부동소수점
print("%f" % 12.34)
# %%표기
print("한국에 산은 %d%% 정도 있습니다." % 70)
```

```python
# 문자열
print("%-20s입니다" % '12.34 문자표현')
# 정수
print("%-10d" % 12.34)
# 부동소수점
print("%f" % 12.34)
# %%표기
print("한국에 산은 %d%% 정도 있습니다." % 70)
```

```python
# 소수점 표현하기
print("%.2fkkkkk" % 3.141592)
print("%10.2fkkkkk" % 3.141592)
print("%-10.2fkkkkk" % 3.141592)
```

### format 함수를 사용한 포맷팅

```python
# 숫자값을 가진 변수로 대입하기
s= "five"
ss = "I eat {0} apples".format(s)
print(ss, type(ss))
```

```python
# 2개 이상 값 넣기
s = "five"
e = 7
ss = "I ate {0} apples, so I was sick for {1} days.".format(s,e)
print(ss, type(ss))
```

```python
# 혼용해서 넣기
sss = "I ate {0} apples, so I was sick for {day} days.".format(10, day=3)
print(sss)
```

```python
# 왼쪽정렬
print("{0:<10}".format("hi"))
# 오른쪽정렬
print("{0:>10}".format("hi"))
# 가운데정렬
print("{0:^10}".format("hi"))
```

```python
# 응용
re = "{0:.^18}".format("영수증")
print("="*20 + "\n" + re + "\n" + "="*20)
```

```python
# input으로 개수를 입력받아
# 한잔에 2500원인 아아와 3800원인 라떼를 계산한 영수증을 출력하라.
ice = input("아아 몇 잔을 주문하시겠습니까?")
latte = input("라떼 몇 잔을 주문하시겠습니까?")
re = "{0:^18}".format("영수증")
print("=" * 20 + "\n" + re + "\n" + "=" * 20)
print("      아아 %s잔" % ice)
print("      라떼 %s잔" % latte)
print("." * 20)
print("    아아 %s원" % (int(ice)*2500))
print("    라떼 %s원" % (int(latte)*3800))
print("-" * 20)
print(">>> 총가격: "+ str(int(ice)*2500 + int(latte)*3800) + " 원" )
print("=" * 20)
```

### 문자열 관련 함수들 

```python
# count - 문자 개수 세기
a = "hobby"
b = "appleppp"
print(a.count('b'))
print(b.count('p'))
```

```python
# find - 위치 알려주기
a = "Python is the best choice"
print(a.find('P'))
print(a.find('b'))
print(a.find('x'))
```

```python
# join - 문자열 삽입
ss = 'abcdefg'
print("-".join(ss))
# a-b-c-d-e-f-g
# print("a","b","c",sep="-")와 다름
```

```python
# join 연습
# a*b*c*d*f*g
ss = 'abcdefg'
print("*".join(ss))
# 1빠 2빠 3빠 ...
nn = '123456789'
print("빠 ".join(nn) + "빠")
```

```python
# upper - 소문자를 대문자로 바꾸기
a = 'Python is the best choice'
print(a.upper())

# lower - 대문자를 소문자로 바꾸기
b = 'Python Is The Best Choice'
print(b.lower())
c = 'Python Is The Best Choice'
print(c[0] + c[1:].lower())
```

```python
ss = f"{'hi':^12}"
print(ss)
# lstrip - 왼쪽 공백지우기
print(ss.lstrip())
# rstrip - 오른쪽 공백지우기
print(ss.rstrip())
# strip - 양쪽 공백지우기
print(ss.strip())
```

```python
# replace - 문자열 바꾸기
a = "Life is too short"
b = a.replace("too short","Egg")
print(b)
print(a)
```

```python
# split - 문자열 나누기
a = "Life is too short"
print(a.split())
b = "a:b:c:d:e:f"
print(b.split(':'))

#BaNaNa -> banana -> b*a*n*a*n*a -> ['b','a','n','a','n',a]
c = "BaNaNa"
d = c.lower()
e = "*".join(d)
f = e.split('*')
print(f)
```

```python
# isalpha - 문자열이 알파벳으로만 구성되어 있는지 확인하기
s = "Life is too short"
print(s.replace(" ","").isalpha()) # 메소드체이닝
```

```python
# isdigit - 문자열이 숫자로만 구성되어 있는지 확인하기
n = "123456"
print(n.isdigit())
```

```python
# startswith - 문자열이 특정 문자(열)로 시작하는지 확인하기
s = "Life is too short"
print(s.startswith("Life"))
print(s.startswith("short"))
```

```python
# endswith - 문자열이 특정 문자(열)로 끝나는지 확인하기
s = "Life is too short"
print(s.endswith("short"))
print(s.endswith("too"))
```
