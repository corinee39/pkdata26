## 여러 개의 입력값을 받는 함수 만들기

```python
def add_mul(choice, *args):
    if choice == "add": # 매개변수 choice에 "add"를 입력받았을 때
        result = 0
        for i in args:
            result = result + i
    elif choice == "mul": # 매개변수 choice에 "mul"를 입력받았을 때
        result = 1
        for i in args:
            result = result * i
    return result
```

```python
result = add_mul('add', 1,2,3,4,5)
print(result)
result = add_mul('mul', 1,2,3,4,5)
print(result)
```

```python
# 위 예제에서 args를 하나만 넣고도 동작하도록 만들어보기
def add_mul(*args):
    result1 = 0
    for i in args:
        result1 = result1 + i
    result2 = 1
    for i in args:
        result2 = result2 * i
    return result1, result2
```

```python
add_mul(1,2,3,4,5)
```

```python
def print_test(*args):
    print(args,type(args))
print_test(1,2,3,4,5)
```

##  매개변수에 초깃값 미리 설정하기 * 

```python
# 매개변수에 초깃값 미리 설정은 함수를 생성할 때 만든다
def say_myself(name, age, man=True):
    print("나의 이름은 %s 입니다." % name)
    print("나이는 %d살입니다." % age)
    if man:
        print("남자입니다.")
    else:
        print("여자입니다.")
say_myself("홍길동",22)
say_myself("박서방",23,man=True)
say_myself("향단이",24,man=False)
```

## lambda 예약어 *

```python
# 사용법
add = lambda a, b : a + b
result = add(11,22)
print(result)
```

```python
# 한 줄짜리 익명 함수
print((lambda x: x**3)(2))
```

```python
# lambda 응용
students = [("철수", 90), ("영희", 85), ("민수", 95)]
ss = sorted(students, key = lambda x: x[1], reverse = True)
print(ss)
```

```python
# json으로 파싱하여 Lambda로 키를 설정하여 정렬하기
import random as ran
import requests as req
url = "https://jsonplaceholder.typicode.com/todos"
result = req.get(url).json()

print("1. id의 번호를 기준으로 내림차순으로 정렬하라.")
fin = sorted(result, key = lambda x : x['id'], reverse = True)
print(fin)
print("="*100)

print("2. completed가 True 된 내용만 전처리하라.")
print ([n for n in fin if n['completed'] == True])
print("="*100)

print("3. 리스트의 각 요소 끝에 score 이름으로 40~100까지의 임의의 값을 배정하여 fix라는 변수에 리스트를 등록하라.")
fix = []
for n in fin:
    score = ran.randint(40,100)
    n["score"] = score
    fix.append(n)
print(fix)
print("="*100)

print("4. score를 키로 내림차순으로 원본손상없이 sort하라.")
fix = sorted(fix ,key = lambda x : x['score'], reverse = True)
print(fix)
print("="*100)
print(fin) # 원본 손상 확인
```
