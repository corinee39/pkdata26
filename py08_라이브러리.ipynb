```python
# 기본코드(1세대 공청기 필터)
test_air = [1,23,32,-5,45,0,-3,123,2,-13]
def pos(num):
    result = []
    for n in num:
        if n > 0:
            result.append(n)
    return result

print(pos(test_air))
```

```python
# 기본코드 응용(리스트컴프리헨션)
test_air = [1,23,32,-5,45,0,-3,123,2,-13]
print([n for n in test_air if n > 0])
```

```python
# filter ** 중요 filter(함수, 반복_가능한_데이터)
# 2세대 공청기용 필터 기본코드 // 필터는 리턴이 True!! 가 중요

test_air = [1,23,32,-5,45,0,-3,123,2,-13]
def pos_fn(n):
    return n > 0
def neg_fn(n):
    return n < 0

print(list(filter(pos_fn, test_air)))
print(list(filter(neg_fn, test_air)))
```

```python
# filter 응용(람다함수)
test_air = [1,23,32,-5,45,0,-3,123,2,-13]
print(list(filter(lambda x: x > 0, test_air)))
print(list(filter(lambda x: x < 0, test_air)))
```

```python
# 기본코드(가공 1)
sample = [1, 2, 3, 4, 11, 22, 33, 44]
def two_times(numberList):
    result = []
    for number in numberList:
        result.append(number * 2)
    return result

result = two_times(sample)
print(result)
```

```python
# 5장 되새김 문제 1번
class Calculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val
        return self.value

class UpgradeCalculator(Calculator):
    def minus(self, val):
        self.value -= val
        return self.value
        
cal = UpgradeCalculator()
cal.add(10)
cal.minus(7)

print(cal.value)
```

```python
# 5장 되새김 문제 2번
class Calculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val
        return self.value

class MaxLimitCalculator(Calculator):
    def add(self, val):
        self.value += val
        if self.value > 100:
            self.value = 100
        else:
            return self.value
        
cal = MaxLimitCalculator()
cal.add(50)
cal.add(60)

print(cal.value)
```

```python
from konlpy.tag import Okt
okt = Okt()

positive_words = {
    "행복하다": 2,
    "사랑하다": 2,
    "감사하다": 2,
    "즐겁다": 2,
    "아름답다": 2,
    "감동하다": 1,
    "만족하다": 1,
    "희망하다": 2,
    "존경하다": 2,
    "기쁘다": 2,
    "뿌듯하다": 1,
    "건강하다": 1,
    "성공하다": 2,
    "믿음직하다": 1,
    "따뜻하다": 1,
    "자랑스럽다": 2,
    "용기있다": 1,
    "긍정적이다": 2,
    "밝다": 1,
    "신나다": 2
}

negative_words = {
    "슬프다": -2,
    "화나다": -2,
    "미워하다": -2,
    "불행하다": -2,
    "지겹다": -1,
    "무가치하다": -2,
    "괴롭다": -2,
    "싫다": -2,
    "두렵다": -1,
    "우울하다": -2,
    "절망하다": -2,
    "불안하다": -1,
    "짜증나다": -1,
    "외롭다": -2,
    "후회하다": -1,
    "비참하다": -2,
    "실망하다": -1,
    "억울하다": -1,
    "초조하다": -1,
    "불편하다": -1
}

text = input("당신의 기분을 한 문장으로 표현해 주세요.")
words = okt.morphs(text, stem = True)

# 리스트 순회하여 키값 얻기
li = []
for x in words:
    try:
        if x in positive_words:
            li.append(positive_words.get(x))
        elif x in negative_words:
            li.append(negative_words.get(x))
    except:
        continue

# 점수 계산
score = 0
for y in li: 
    score += y

if score >= 0:
    print(f"기분이 좋아보이네요. 당신의 점수는 {score}입니다.")
else:
    print(f"기분이 나빠보이네요. 당신의 점수는 {score}입니다.")

# 한글 형태소 분석기
# print(words)
```

```python
# functools.reduce
def add(data):
    result = 0
    for i in data:
        result += i
    return result
data = [1, 2, 3, 4, 5]
result = add(data)
print(result)
```
