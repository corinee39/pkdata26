# 내장함수

```python
# abs
print(abs(3))
print(abs(-3))
print(abs(-1.3))
```

```python
# all *
# 단, all의 입력 인수가 빈 값인 경우에는 True를 반환한다.
print(all([1,2,3,4,5]))
print(all([1,2,3,'안녕','반가워']))
print(all([1,'',3,4]))
print(all([]))
```

```python
# any *
# 단, any의 입력 인수가 빈 값인 경우에는 False를 반환한다.
print(any([1,2,3,4,5]))
print(any([False,2,0,'']))
print(any(['안녕','하세요','!',0]))
print(any([]))
```

```python
# 아스키코드(ASCII) - 정의: 7비트(128개 문자)
# 유니코드 - 정의: 전 세계 모든 문자를 표현할 수 있는 문자코딩, UTF-8(인코딩방식)
print(chr(97))
print(chr(44033))
print(ord('a'))
print(ord('각'))
print(chr(int(hex(33)[2:], 16))) # 다양하게 엮은 함수들
```

```python
from konlpy.tag import Okt

# 한글 형태소 분석기
okt = Okt()
text = input("당신의 기분을 한 문장으로 표현해 주세요.")
words = okt.morphs(text, stem = True)
print(words)
```
