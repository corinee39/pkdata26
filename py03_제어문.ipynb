```python
# 로또 번호
# 1 ~ 45 임의의 수 중에서 중복없는 6개의 수를 뽑아 나열하여 분석하시오.
import random as ran
n = 0
lotto = []

while n < 6:
    num = ran.randint(1,45)
    # 중복검사
    if num in lotto:
        pass    
    else:
        n += 1
        lotto.append(num)
    
lotto.sort()
print(lotto)
```

```python
# 로또 번호 심화
# 1 ~ 45 임의의 수 중에서 중복없는 6개의 수를 뽑아 나열하여 분석하시오.
# while 반복해서 1000이상으로 돌려서 가장 많이 출현한 숫자를 뽑으세요.
from collections import Counter
import random as ran

lotto_list = []
fre = 0

while fre < 1000:
    
    # 6개의 수 뽑기
    n = 0
    lotto = []
    while n < 6:
        num = ran.randint(1,45)
        
        # 중복검사
        if num in lotto:
            pass    
        else:
            n += 1
            lotto.append(num)
            lotto_list.append(num)
    lotto.sort()
    # print(lotto)
    fre += 1    
# print(lotto_list)

# 최다 출현 숫자
lotto_count = Counter(lotto_list)
# print(lotto_count)
max_num = max(lotto_count, key = lotto_count.get)
max_count = max(lotto_count.values())
# print(max_num, max_count)
print(f"가장 많이 출현한 숫자는 {max_num}이고, 그 횟수는 {max_count}회입니다.")
```

```python
# 주사위 눈금 조작_3번을 16.x -> 25%로 조작하라
import random as ran

box = []
n = int(input("주사위를 몇번 던지시겠습니까?") or 100) # 숫자를 넣지 않으면 100
lucky = int(input("조작할 확률은 몇 %입니까?") or 10) # 퍼센트 조절
my = int(input("조작할 번호는 몇 번입니까?"))

while len(box) < n:
    num = ran.randint(1,6) # 1~6까지 숫자 중 랜덤
    if (num != my) and (((box.count(my)/n)*100) < lucky):
        pass
    elif (num == my) and (((box.count(my)/n)*100) > lucky):
        pass
    else:
        box.append(num)

if n == len(box):
    print(f"{n}회 정상 진행되었습니다.")
    for num in range(1,7):
        print(f"주사위 눈금 {num}이 {box.count(num)}번 {((box.count(num)/n)*100):0.2f}% 출현하였습니다.")
else:
    print(f"검증 결과 이상이 있습니다. {n}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")
```

```python
import random as ran
import matplotlib.pyplot as plt
box = []
n = 0
cc = 10000
values = []
choice = int(input("조작할 확률은 몇 %입니까?") or 10) # 퍼센트 조절
my = int(input("조작할 번호는 몇 번입니까?") or 3)
while n < cc:
    num = ran.randint(1,6)
    lucky = ran.randint(1,100)
    if choice >= lucky:
        box.append(my)
    else:
        box.append(num)
    n += 1
    
if cc == len(box):
    print(f"{cc}회 정상 진행되었습니다.")
    for n in range(1,7):
        print(f"주사위 눈금 {n}이 {box.count(n)}번 {((box.count(n)/cc)*100):0.2f}% 출현하였습니다.")
        values.append(box.count(n))
else:
    print(f"검증 결과 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")

color = ["skyblue","orange","skyblue","orange","skyblue","orange"]
plt.bar(range(1,7), values, color=color)
plt.show()
```

```python
numbers = [1, 2, 3, 4, 5]
result = []
for n in numbers:
    if n % 2 == 1:
        result.append(n * 2)
print(result)
```

```python
numbers = [1, 2, 3, 4, 5]
result = [n * 2 for n in numbers if n % 2 == 1]
print(result)
```
