### 리스트 관련함수 

```python
# append - 추가하기
a = []
print(a)
a.append(1)
print(a)
a.append(2)
print(a)
a.append([11,22,33])
print(a)
a.append(4)
print(a)
```

```python
# sort - 리스트 정렬
a = [3,4,6,2,34,23,76,3,4]
b = ['m','ds','w','er','f','e','e','ks']
a.sort()
b.sort()
print(a)
print(b)
```

```python
# reverse - 리스트 뒤집기
a = [3,4,6,2,34,23,76,3,4]
b = ['m','ds','w','er','f','e','e','ks']
print(a[::-1]) # 원본이 바뀌지 않음
print(b[::-1])
a.reverse()
b.reverse()
print(a) # 원본이 바뀜
print(b)
```

```python
# sort 와 reverse 응용
a = [3,4,6,2,34,23,76,3,4]
a.sort()
a.reverse() # a[::-1]
print(a)

b = [3,4,6,2,34,23,76,3,4]
b.sort(reverse=True) # True는 내림차순, False 오름차순이 기본값
print(b)
```

```python
# sorted - 원본 유지하기
aa = [66,7,88,9,0]
print(sorted(aa))
print(sorted(aa,reverse=True))
print(aa)
```

```python
# sorted 응용하기
fruit = ['banana', 'kiwi', 'orange', 'melon', 'apple', 'strawberry']
print(sorted(fruit)) 
print(sorted(fruit,key=len)) # 글자 길이의 오름차순
print(sorted(fruit,key=len,reverse=True)) # 글자 길이의 내림차순
```

### 딕셔너리 관련함수

```python
# keys - 키 리스트 만들기
a = {"name": "pey", "phone": "010-9999-1234", "birth": "1118"}
ky = a.keys()
l_ky = list(ky)
print(ky,type(a),type(ky),type(l_ky),sep="\n")
```

```python
# values - 값 리스트 만들기
a = {"name": "pey", "phone": "010-9999-1234", "birth": "1118"}
va = a.values()
l_va = list(va)
print(va,type(a),type(va),type(l_va),sep="\n")
```

```python
# items - 키값쌍을 리스트 안의 튜플로 만들기
a = {"name": "pey", "phone": "010-9999-1234", "birth": "1118"}
it = a.items()
l_it = list(it)
print(it,type(a),type(it),type(l_it),sep="\n")
```

```python
# clear - 키값쌍 모두 지우기
a = {"name": "pey", "phone": "010-9999-1234", "birth": "1118"}
print(a)
a.clear()
print(a)
```

```python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
# 교집합 &
print("교집합", s1 & s2) # s1.intersection(s2) 가능
# 합집합 |
print("합집합", s1 | s2) # s1.union(s2) 가능
# 차집합 -
print("차집합", s1 - s2) # s1.difference(s2) 가능
```

### 집합 자료형 관련 함수

```python
# add - 집합자료형 값 1개 추가
s1 = set([1,2,3])
print(s1)
s1.add(4)
print(s1)
```

```python
# update - 집합자료형 값 여러개 추가
s1 = set([1,2,3])
print(s1)
s1.update([2,3,4,5,6,7])
print(s1)
```

```python
# remove - 집합자료형 특정값 제거
s1 = set([1,2,3])
print(s1)
s1.remove(1) # index가 아닌 값으로 선택
print(s1)
```

**존재하지 않는 값을 찾을 땐 오류**<br/>


```python
# discard- 집합자료형 특정값 제거
s1 = set([1,2,3])
print(s1)
s1.discard(0) # 없는 값이어도 오류가 발생하지 않음
print(s1)
```

```python
# clear- 집합자료형 내 모든값 제거
s1 = set([1,2,3])
s1.clear()
print(s1)
```

```python
# or 연산자
print("True or True ->",True or True)
print("True or False ->",True or False)
print("False or True ->",False or True)
print("False or False ->",False or False)
```

```python
# not 연산자
print("not True ->",not True)
print("not False ->",not False)
print("not 1 ->",not 1)
print("not 0 ->",not 0)
```

```python
print(bool("안녕"),bool(""))
print(not bool("안녕"),not bool(""))
print(not not bool("안녕"),not not bool(""))
```

```python
# id 메모리 주소 확인
a = [1,2,3]
aa = a  # 참조O / 복사X
b = [1,2,3]
print(id(a))
print(id(aa))
print(id(b))
```

```python
print(aa == a) # 값만을 비교
print(a == b)
print(aa is a) # 뿌리까지 비교
print(a is b)
```

```python
e = [11,22,33]
ee = e
ee[2] = 77
print(ee is e, ee, e)
```

```python
e = [11,22,33]
ee = e[:]
ee[2] = 77
print(ee is e, ee, e)
```

```python
e = [11,22,33]
ee = e.copy()
ee[2] = 77
print(ee is e, ee, e)
```

### 변수를 만드는 여러가지 방법 

```python
(a, b, c) = (123, 456, 789)
print(a)
print(b)
print(c)
```

```python
[a,b,c] = ['python', 'life', 'cool']
print(a)
print(b)
print(c)
```

```python
a = b = c = 0
print(a)
print(b)
print(c)
```

```python
# 두 변수의 값을 바꿀때
a = 33
b = 55
print(a, b)

a, b = b, a # 파이썬에서만 가능
print(a, b)
```

```python
# 다른 언어 예 -> temp 사용
aa = 33
bb = 55
print(aa, bb)

temp = aa
aa = bb
bb = temp
print(aa, bb)
```
