```python
def counter_maker():
    count = [0] # 리스트로 감싸서 mutable하게 만듦

    def counter():
        count[0] += 1
        return count[0]

    return counter

c1 = counter_maker()
c2 = counter_maker()
```

```python
c1()
```

```python
c2()
```

```python
print(c1.__closure__[0].cell_contents)
print(c2.__closure__[0].cell_contents)
```

## 데코레이터 

```python
# 데코레이터
import time

def myfunc():
    start = time.time()
    print("함수가 실행됩니다.")
    print("함수 수행시간: %f 초" % (time.time() - start))
myfunc()
```

```python
import time

def outfunc(X):
    def infunc():
        start = time.time()
        print("함수 준비")
        result = X()
        print("함수 수행시간: %f 초" % (time.time() - start))
        return result
    return infunc

@outfunc
def Xfunc():
    print(" 함수가 실행됩니다.")
@outfunc
def Yfunc():
    print(" 함수가 어렵습니다.")
```

```python
Xfunc()
Yfunc()
```

```python
Xfunc = outfunc(Xfunc)
Yfunc = outfunc(Yfunc)
Xfunc()
Yfunc()
```

```python
# 이터레이터 만들기
class MyIterator:
    def __init__(self, data):
        self.data = data
        self.position = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.position >= len(self.data):
            raise StopIteration
        result = self.data[self.position]
        self.position += 1
        return result

# if __name__ == "__main__":
#     i = MyIterator([1, 2, 3])
#     for item in i:
#         print(item)
```

```python
it = MyIterator([1, 2, 3])
```

```python
for n in it:
    print(n)
```

```python
# 제너레이터
def mygen():
    yield 1
    yield 2
    yield 3

g = mygen()
```

```python
next(g)
```

```python
def mybot():
    print("작업 시작")
    yield "1번 작업완료"
    print("로봇A 위치로 동작")
    yield "2번 작업완료"
    print("로봇A 그리퍼 동작 후 위치반환")
    yield "3번 작업완료"
    print("로봇B 가동 시작")
    yield "end"

gb = mybot()
```

```python
try:
    next(gb)
except StopIteration as e:
    print("더이상 할 작업이 없습니다.", e)
```
