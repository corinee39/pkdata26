```python
# 응용

import requests as req
from bs4 import BeautifulSoup as bs
url = "https://www.melon.com/chart/index.htm"
headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36 Edg/145.0.0.0"}
web = req.get(url, headers=headers)
web.content
soup = bs(web.content, 'html.parser')

title = soup.select('.rank01')[:20]
name = soup.select('.checkEllipsis a')[:20]
str = ''
for i, (t, n) in enumerate(zip(title, name), 1):
    str += f'{i}위: {t.text.strip()} / {n.text} \n'
print(str,end='')
```

```python
# 클로저 *  함수를 리턴

def calc():
    a = 3
    b = 5
    def mul_add(x):
        return a * x + b
    return mul_add
```

```python
def counter_maker():
    count = [0] # 리스트로 감싸서 mutable하게 만듦

    def counter():
        count[0] += 1
        return count[0]

    return counter

c1 = counter_maker()
c2 = counter_maker()
```

```python
c1()
```

```python
c2()
```

```python
print(c1.__closure__[0].cell_contents)
print(c2.__closure__[0].cell_contents)
```

## 데코레이터 

```python
# 데코레이터
import time

def myfunc():
    start = time.time()
    print("함수가 실행됩니다.")
    print("함수 수행시간: %f 초" % (time.time() - start))
myfunc()
```

```python
import time

def outfunc(X):
    def infunc():
        start = time.time()
        print("함수 준비")
        result = X()
        print("함수 수행시간: %f 초" % (time.time() - start))
        return result
    return infunc

@outfunc
def Xfunc():
    print(" 함수가 실행됩니다.")
@outfunc
def Yfunc():
    print(" 함수가 어렵습니다.")
```

```python
Xfunc()
Yfunc()
```

```python
Xfunc = outfunc(Xfunc)
Yfunc = outfunc(Yfunc)
Xfunc()
Yfunc()
```
