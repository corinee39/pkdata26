```python
# 비트코인 API
import requests as req
url = "https://api.bithumb.com/v1/candles/minutes/1?market=KRW-BTC&count=1"
d_url = "https://api.bithumb.com/v1/candles/days?market=KRW-BTC&count=1"
ori = req.get(url).json()[0] # list -> dict
ori2 = req.get(d_url).json()[0] 

#'''분봉'''
# 시작가
openp = float(ori['opening_price'])
# 최고가
maxp = float(ori['high_price'])
# 최저가
minp = float(ori['low_price'])
# 변화폭
bandp = maxp - minp

#'''일봉'''
# 시작가
openp2 = float(ori2['opening_price'])
# 최고가
maxp2 = float(ori2['high_price'])
# 최저가
minp2 = float(ori2['low_price'])
# 변화폭
bandp2 = maxp2 - minp2

print("분기준 시작가:",openp,"\n최고가:",maxp,"\n최저가:",minp,"\n변화폭:",bandp)
print("일기준 시작가:",openp2,"\n최고가:",maxp2,"\n최저가:",minp2,"\n변화폭:",bandp2)
# print("검토:",openp + bandp)
# print("최고가:",maxp)

# 파이썬 외 언어, 삼항연산자 ((maxp - (openp + bandp)) > 0)? "떡상중" : "떡락중"
result = "떡상중" if ((maxp - (openp + bandp)) > 0) else "떡락중"
result2 = "떡상중" if ((maxp2 - (openp2 + bandp2)) > 0) else "떡락중"
print("분봉으로 판단:",result)
print("일봉으로 판단:",result2)
```

```python
import requests as req

# 검색하려는 지역의 위도, 경도 입력받기
lat = input("날씨를 알고 싶은 지역의 위도를 입력하세요.")
lon = input("날씨를 알고 싶은 지역의 경도를 입력하세요.")

url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current_weather=true"
res = req.get(url).json()

# 날씨 단위
unit_date = res['current_weather_units']['temperature']
unit_ws = res['current_weather_units']['windspeed']

# 지역 날씨 정보
date = res['current_weather']['time'][:10]
time = res['current_weather']['time'][11:]
temp = res['current_weather']['temperature']
ws = res['current_weather']['windspeed']
print(f"{date} GMT+0 {time}시 기준으로 해당 지역은 기온 {temp}{unit_date}, 풍속 {ws}{unit_ws} 입니다.")
```

```python
import requests as req

url = "https://api.fda.gov/drug/event.json?count=patient.reaction.reactionmeddrapt.exact"
res = req.get(url).json()

# 구토(NAUSEA) 증상의 횟수
cnt_NAU = res['results'][3]['count']
print("1. 구토 증상의 횟수는 " + f"{cnt_NAU:,}" + "회입니다")

# COVID-19 보고 건수
cnt_COV = res['results'][53]['count']
print("2. COVID-19 보고 건수는 " + f"{cnt_COV:,}" + "회입니다")

# 심근경색 증상 횟수
cnt_MYO = res['results'][54]['count']
print("3. 심근경색 증상의 횟수는 " + f"{cnt_MYO:,}" + "회입니다")

# COVID-19과 심근경색 비교
comp = "코로나가 더 많음" if (cnt_COV > cnt_MYO) else "심근경색이 더 많음"
print("4. 비교 결과:", comp)

# NAUSEA와 VOMITING 비교
cnt_VOM = res['results'][10]['count']
comp2 = "NAUSEA" if (cnt_NAU > cnt_VOM) else "VOMITING"
sub = cnt_NAU - cnt_VOM if (cnt_NAU > cnt_VOM) else cnt_VOM - cnt_NAU
print("5. 보고 건수는 ",comp2,"가 더 많으며 차이는",f"{sub:,}","회입니다")

# "FDA" 단어 포함 여부
con = "FDA" in res['meta']['disclaimer']
con2 = "6. 포함되어 있습니다" if (con == True) else "6. 포함되어있지 않습니다"
print(con2)
```
